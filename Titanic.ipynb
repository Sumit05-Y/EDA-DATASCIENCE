```python
import pandas as pd


url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)
print(f"Dataset shape: {df.shape}")
df.info()
```

```python
numeric_df = df.select_dtypes(include='number')
print(numeric_df)
```

```python
survivors_over_30=df[(df['Age']>30)&(df['Survived']==1)]
print(f"Survivors over 30: {len(survivors_over_30)}")
survivors_over_30.head()
```

```python
survival_by_class= df.groupby("Pclass")['Survived'].mean()
print(survival_by_class)
```

```python
for pclass,rate in survival_by_class.items():
  print(f"Class {pclass}: {rate:.2%} Survived")
```

```python
subset = df[["Pclass","Sex","Age","Fare","Survived"]].copy()
subset.to_csv('titanic_subset.csv',index = False)
print("Saved")

from google.colab import files
files.download("titanic_subset.csv")
```
